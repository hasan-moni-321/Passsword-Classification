### Tokenizing

```python
def word_to_character(inputs):
    characters = []
    for i in inputs:
        characters.append(i)
        
    return characters
```

```python
vectorizer = TfidfVectorizer(tokenizer=word_to_character)
X = vectorizer.fit_transform(X)
```

```python
X.shape
```

```python
vectorizer.vocabulary_
```

### Printing first element of the vectorizer

```python
data.iloc[0,0]
```

```python
feature_name = vectorizer.get_feature_names()

#get tfidf vector for first document
first_document_vector=X[0]

#printing for the first document
df_first = pd.DataFrame(first_document_vector.T.todense(), index=feature_name, columns=['tfidf'])
df_first.sort_values(by=['tfidf'], ascending=False)
```

### Dividing dataset to Feature and label data

```python
Xtrain,xtest, Ytrain,ytest = train_test_split(X,y, test_size=.25, random_state=42)
```

### Define Algorithm

```python
lr = LogisticRegression(solver='lbfgs', multi_class='auto',  max_iter=1000)
lr.fit(Xtrain,Ytrain)
```

```python
lr.score(xtest, ytest)
```

```python
## With Multinomial
li_re = LogisticRegression(random_state=0, multi_class='multinomial', solver='newton-cg')
li_re.fit(Xtrain,Ytrain)
```

```python
li_re.score(xtest, ytest)
```

### Predicting with new data

```python
new_data = np.array(["hsdmon@ow13"])
vector = vectorizer.transform(new_data)
pred = li_re.predict(vector)
```
